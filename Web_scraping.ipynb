```python
import requests

url = "https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_6_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_6_0_na_na_na&as-pos=6&as-type=TRENDING&suggestionId=mobiles&requestId=acf3ca05-4f9e-4dca-8034-924059566256&page=1"
response = requests.get(url)

print("Response Status : ", response.status_code)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text, "html.parser")
print(soup.prettify())  # Beautifies HTML output
```

```python
title = soup.find("title").text
print("Title:", title)

links = soup.find_all("a")  # Get all links
for link in links:
    print()
    print(link.get("href"))
```

```python
product_name = []
names = soup.find_all("div",class_= 'KzDlHZ')
for i in names:
    product_name.append(i.text)
print(product_name)
```

```python
product_price = []
price = soup.find_all("div",class_= 'Nx9bqj _4b5DiR')
for i in price:
    product_price.append(i.text)
print(product_price)
```

```python
p_review = []
review = soup.find_all("div",class_= 'XQDdHH')
for i in review:
    p_review.append(i.text)
print(p_review)

product_review = p_review[0:len(product_name)]
```

```python
product_discount = []
discount = soup.find_all("div",class_= 'UkUFwK')
for i in discount:
    product_discount.append(i.text)
print(product_discount)

product_discount = product_discount[0:len(product_name)]
```

```python
import pandas as pd
```

```python
data = {
    "product_name" : product_name,
    "product_price" : product_price,
    "product_review" : product_review,
    "product_discount" : product_discount
}
```

```python
df = pd.DataFrame(data)
```

```python
df.head(3)
```

```python
df["product_price"] = df["product_price"].str.replace(",", "")
df["product_price"] = df["product_price"].str.replace("₹", "")
df["product_price"] = df["product_price"].astype(int)
```

```python
df["product_review"] = df["product_review"].astype(float)
```

```python
df["product_discount"] = df["product_discount"].str.replace("% off", "")
df["product_discount"] = df["product_discount"].astype(int)
```
