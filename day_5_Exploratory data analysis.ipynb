# EDA ===> Exploratory Data analysis

## Parts of EDA
* 1. Univariate Analysis ==> Analysis on a single Independent column
* 2. Bivariate Analysis ==> Analysis on two columns
* 3. Multivariate Analysis Analysis on more than 2 columns

### Data Types
* 1. Numerical Data ===> Continuous data ===> Age(year ,month, days), Height, weight etc..
* 2. Categorical Data ===> Descrete data ===> Total Number of Employees

# Tips Dataset

```python
tips = sns.load_dataset("tips")
tips
```

# Bivariate Analysis

```python
# 1. scatterplot(Numerical cloumn - Numerical column)
```

```python
sns.scatterplot(x = tips["total_bill"], y = tips["tip"])
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips, hue=tips["sex"])
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips, hue=tips["time"])
```

```python
sns.scatterplot(x="total_bill", y="tip", data=tips, hue=tips["sex"], style = tips["smoker"])
```

```python
# HeatMap (Categorical - categorical)
p = pd.crosstab(tips["day"], tips["time"])
p
```

```python
sns.heatmap(p)
```
