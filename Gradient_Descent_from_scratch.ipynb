```python
from sklearn.datasets import make_regression
import numpy as np
```

# n samples = 4: This specifies the number of samples (rows) in
# ttie dataset. In this case, the dataset will have four rows.
# n _ features = I: This specifies the number of features (columns) in the dataset. In this case, the dataset will have one feature
# column.
# n informative = I: This specifies the number of informative features in the dataset. Informative features are those that are
# used to generate the target variable. In this case, there is only one informative feature column.
# n_targets = 1: This specifies the number of target variables in the dataset. In this case, there is only one target variable.
# noise = 80: This specifies the amount of noise to add to the target variable. In this case, the noise level is quite high, as it is set
# to 80.**bold text**

```python
x , y = make_regression(n_samples=4,n_features = 1,
                        n_informative=1,
                        n_targets=1,
                        noise = 80 , random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
# lets apply OLS
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

# copy_X: This parameter specifies whether the data set, X,
should be copied or not before fitting the linear regression
model. By default, it is set to True, which means a copy of X
will be created.

```python
LinearRegression(copy_X=True,
                 fit_intercept=True,
                 n_jobs=None)
```

```python
reg.coef_ # m slope
```

```python
reg.intercept_ #b slope
```

```python
plt.scatter(x,y)
plt.plot(x , reg.predict(x),color = 'red')
```

```python
# lets apply Gradient Descent assuming slope is
# constant = 78.35
# and Lets assume the starting value for
# intercept b=0
y_pred = ((78.35*x)+0).reshape(4)  #y=m*x+b
```

```python
plt.scatter(x,y)
plt.plot(x, reg.predict(x), color="red", label="ols")
plt.plot(x, y_pred, color="green", label="b=0")
plt.legend()
plt.show()
```

```python
m=78.35
b = 0
loss_slope = -2*np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
# Lets take learning rate = 0.1

lr = 0.1
step_size = loss_slope*lr
step_size
```

```python
# calculating the new intercept
b = b-step_size   # b_new = b_old - learning_rate +dl/db
b
```

```python
y_pred1 = ((78.35*x)+b).reshape(4)
plt.scatter(x,y)
plt.plot(x, reg.predict(x), color="black", label="ols")
plt.plot(x, y_pred1, color="red", label="b = {}".format(b))
plt.plot(x, y_pred, color="yellow", label="b=0")
plt.legend()
plt.show()
```

```python
# Iteration 2
loss_slope = -2*np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b- step_size
b
```

```python
y_pred2 = ((78.35*x)+b).reshape(4)
plt.scatter(x,y)
plt.plot(x, reg.predict(x), color="black", label="ols")
plt.plot(x, y_pred2, color="green", label="b = {}".format(b))
plt.plot(x, y_pred1, color="red", label="b = {}".format(b))
plt.plot(x, y_pred, color="yellow", label="b=0")
plt.legend()
plt.show()
```

```python
# Itertion 3
loss_slope = -2*np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b-step_size
b
```

```python
y_pred3 = ((78.35*x)+b).reshape(4)
plt.scatter(x,y)
plt.plot(x, reg.predict(x), color="black", label="ols")
plt.plot(x, y_pred3, color="brown", label="ols")
plt.plot(x, y_pred2, color="green", label="b = {}".format(b))
plt.plot(x, y_pred1, color="red", label="b = {}".format(b))
plt.plot(x, y_pred, color="yellow", label="b= 0")
plt.legend()
plt.show()
```

```python
# 4th Iteration
loss_slope = -2*np.sum(y-m*x.ravel()-b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b-step_size
b
```

```python
y_pred4 = ((78.35*x)+b).reshape(4)
plt.scatter(x,y)
plt.plot(x, reg.predict(x), color="black", label="ols")
plt.plot(x, y_pred4, color="purple", label="ols")
plt.plot(x, y_pred3, color="brown", label="ols")
plt.plot(x, y_pred2, color="green", label="b = {}".format(b))
plt.plot(x, y_pred1, color="red", label="b = {}".format(b))
plt.plot(x, y_pred, color="yellow", label="b=0")
plt.legend()
plt.show()
```
