```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("D:\\DataSet\\covid_toy.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer()
```

# LabelEncoding

```python
df1 = df.copy()
df1.head(3)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df1["gender"] = le.fit_transform(df1["gender"])
df1["cough"] = le.fit_transform(df1["cough"])
df1["city"] = le.fit_transform(df1["city"])
df1["has_covid"] = le.fit_transform(df1["has_covid"])
```

```python
df1.head(3)
```

# OrdinalEncoding

```python
df2 = df.copy()
df2.head(3)
```

```python
df2 = df2.drop(columns = ["age", "fever"])
df2.head(3)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories = [
    ["Male","Female"], 
    ["Mild","Strong"], 
    ["Kolkata","Bangalore","Delhi","Mumbai"], 
    ["No","Yes"]
])
```

```python
df2_new = oe.fit_transform(df2)
```

```python
df2_new_df = pd.DataFrame(df2_new, columns = df2.columns)
df2_new_df.sample(5)
```
